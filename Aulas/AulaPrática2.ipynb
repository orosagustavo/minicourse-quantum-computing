Simulando uma Porta NOT Quântica (Pauli-X):


Conceito: Analogia entre portas lógicas clássicas e quânticas (Pauli-X como NOT).

Mostrar que a porta Pauli-X quântica atua como a porta NOT clássica.

Inicializar um qubit em $|0\rangle$, aplicar qml.PauliX, e verificar que o estado se torna $|1\rangle$.

Repetir para o estado inicial $|1\rangle$ para mostrar que se torna $|0\rangle$.

```python
dev = qml.device("default.qubit", wires=1) # Shots=1 para estado determinístico

@qml.qnode(dev, shots=1)
def not_circuit(initial_state):
    if initial_state == 1: # Para simular um estado inicial |1>
        qml.PauliX(wires=0)
    qml.PauliX(wires=0) # Aplica a porta NOT
    return qml.sample(qml.PauliZ(0))

# Teste com estado inicial |0> (representado como 0)
result_0 = not_circuit(0)
print(f"NOT de 0 resulta em: {result_0}") # Deve ser -1 (representando |1>)

# Teste com estado inicial |1> (representado como 1)
result_1 = not_circuit(1)
print(f"NOT de 1 resulta em: {result_1}") # Deve ser 1 (representando |0>)

```
